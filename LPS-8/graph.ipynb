```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
data = pd.read_csv('activity_selection_times.csv')

# Create the plot
plt.figure(figsize=(10, 6))

# Plot actual times
plt.plot(data['Size'], data['Time (microseconds)'], 'b-o', label='Actual Time')

# Plot theoretical O(n^2) complexity
sizes = np.array(data['Size'])
theoretical = sizes**2
# Scale theoretical curve to match actual times
scale_factor = np.mean(data['Time (microseconds)']) / np.mean(theoretical)
plt.plot(sizes, theoretical * scale_factor, 'r--', label='O(n) Theoretical')

plt.title('Activity Selection Sort Time Complexity')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (microseconds)')
plt.grid(True)
plt.legend()

plt.show()
```
