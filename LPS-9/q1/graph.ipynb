```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
data = pd.read_csv('nqueens_times.csv')

# Create figure with two subplots (linear and log scale)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Create theoretical data points for O(n!) complexity
x = np.linspace(min(data['Board Size']), max(data['Board Size']), 100)
y_theoretical = np.exp(x * np.log(x))  # Approximation of n!
scale_factor = np.mean(data['Time (microseconds)']) / np.mean(y_theoretical)
y_theoretical = y_theoretical * scale_factor

# Linear scale plot
ax1.scatter(data['Board Size'], data['Time (microseconds)'], 
           c='blue', label='Actual Measurements')
ax1.plot(x, y_theoretical, 'r--', label='Theoretical ~O(n!)')
ax1.set_title('N-Queens Time Complexity\n(Linear Scale)')
ax1.set_xlabel('Board Size (n)')
ax1.set_ylabel('Time (microseconds)')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend()

# Log scale plot
ax2.scatter(data['Board Size'], data['Time (microseconds)'], 
           c='blue', label='Actual Measurements')
ax2.plot(x, y_theoretical, 'r--', label='Theoretical ~O(n!)')
ax2.set_yscale('log')
ax2.set_title('N-Queens Time Complexity\n(Log Scale)')
ax2.set_xlabel('Board Size (n)')
ax2.set_ylabel('Time (microseconds) - Log Scale')
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend()

# Add algorithm description
plt.figtext(0.02, 0.02, 
           "N-Queens with Position Constraints:\n" +
           " Board Size: nn\n" +
           " Position constraint: (row + col) % 3  0\n" +
           " Backtracking algorithm\n" +
           " Time complexity: O(n!)\n" +
           " Space complexity: O(n)",
           bbox=dict(facecolor='white', alpha=0.8, edgecolor='black'),
           fontsize=10)

# Adjust layout
plt.tight_layout()

# Save the plot
plt.savefig('../q1.png', 
            dpi=300, 
            bbox_inches='tight',
            facecolor='white')
```
