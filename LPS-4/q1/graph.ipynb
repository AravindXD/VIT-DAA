```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('fibonacci_results.csv')

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Aravind N 22BRS1099', fontsize=24, fontweight='bold')


# Plot the actual runtime
ax1.plot(df['N'], df['Time (microseconds)'], marker='o')
ax1.set_title('Actual Runtime')
ax1.set_xlabel('Input Size (N)')
ax1.set_ylabel('Time (microseconds)')
ax1.set_yscale('log')  # Use log scale for better visualization
ax1.grid(True)

# Calculate and plot the ideal complexity (O(2^n))
n_values = np.array(df['N'])
ideal_times = 2 ** n_values
ideal_times = ideal_times * (df['Time (microseconds)'].max() / ideal_times.max())

ax2.plot(n_values, ideal_times, label='Ideal (O(2^n))', linestyle='--')
ax2.set_title('Ideal Runtime (O(2^n))')
ax2.set_xlabel('Input Size (N)')
ax2.set_ylabel('Time (arbitrary units)')
ax2.set_yscale('log')  # Use log scale for better visualization
ax2.grid(True)

plt.tight_layout()
plt.show()

# Print the results
print(df)
```
