```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
data = pd.read_csv('matrix_chain_max_times.csv')

# Plot actual time graph
plt.figure(figsize=(10, 6))
plt.plot(data['Size'], data['Time (seconds)'], marker='o', label='Actual Time')
plt.title('Matrix Chain Max Multiplication Time vs Size ARAVIND N 22BRS1099')
plt.xlabel('Size')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.legend()
plt.show()

# Ideal time complexity: O(n^3)
sizes = np.arange(2, 11)
ideal_times = sizes**3

# Plot ideal time complexity graph
plt.figure(figsize=(10, 6))
plt.plot(sizes, ideal_times, linestyle='--', color='red', label='Ideal O(n^3)')
plt.title('Ideal Time Complexity for Matrix Chain Max Multiplication ARAVIND N 22BRS1099')
plt.xlabel('Size')
plt.ylabel('Time (arbitrary units)')
plt.grid(True)
plt.legend()
plt.show()
```
