```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('catalan_times.csv')

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(data['Size'], data['Time (seconds)'], marker='o')
plt.title('Catalan Number Calculation Time vs Size ARAVIND N 22BRS1099')
plt.xlabel('Size')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.savefig('catalan_times_plot.png')  # Save the plot as a file
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate sizes
sizes = np.arange(2, 21)

# Simulate ideal time complexity (O(n^2))
ideal_times = sizes**2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sizes, ideal_times, label='Ideal O(n^2)', linestyle='--', color='red')
plt.title('Ideal Time Complexity for Catalan Number Calculation ARAVIND N 22BRS1099')
plt.xlabel('Size')
plt.ylabel('Time (arbitrary units)')
plt.grid(True)
plt.legend()
plt.show()
```
