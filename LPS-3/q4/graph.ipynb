```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
df = pd.read_csv('results_maxsubarray.csv')

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Actual Data
ax1.plot(df['Size'], df['Time (microseconds)'], 'b-', label='Actual')
ax1.set_xlabel('Input Size')
ax1.set_ylabel('Time (microseconds)')
ax1.set_title('Actual Performance: Size vs Time ARAVIND N 22BRS1099')
ax1.legend()
ax1.grid(True)

# Use log scale for better visualization
ax1.set_xscale('log')
ax1.set_yscale('log')

# Plot 2: Idealistic Comparison
# Generate points for ideal O(n log n) complexity
x = np.linspace(min(df['Size']), max(df['Size']), 100)
y = x * np.log(x)  # For O(n log n)

ax2.plot(x, y, 'r-', label='Ideal O(n log n)')
ax2.plot(df['Size'], df['Time (microseconds)'], 'b-', label='Actual')
ax2.set_xlabel('Input Size')
ax2.set_ylabel('Time (arbitrary units)')
ax2.set_title('Ideal vs Actual: Size vs Time')
ax2.legend()
ax2.grid(True)

# Use log scale for better visualization
ax2.set_xscale('log')
ax2.set_yscale('log')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()

# Calculate and print the slope of the actual data (in log-log scale)
log_size = np.log(df['Size'])
log_time = np.log(df['Time (microseconds)'])
slope, _ = np.polyfit(log_size, log_time, 1)
print(f"Approximate experimental time complexity: O(n^{slope:.2f})")

# Calculate and print the ratio of actual time to n log n
n_log_n = df['Size'] * np.log(df['Size'])
ratio = df['Time (microseconds)'] / n_log_n
avg_ratio = np.mean(ratio)
print(f"Average ratio of actual time to n log n: {avg_ratio:.2f}")
```
